```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Read image
img = cv2.imread("test2.png")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Step 2: Convert to binary
_, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

# Step 3: Connected Component Analysis
num_labels, labels, stats, centroids = cv2.connectedComponentsWithStats(binary, connectivity=8)

# Step 4: Create colored output image
output = np.zeros_like(img)

# Assign random colors to each label
for label in range(1, num_labels):  # skip background (label 0)
    mask = labels == label
    color = np.random.randint(0, 255, size=3)
    output[mask] = color

    # Get centroid
    x, y = int(centroids[label][0]), int(centroids[label][1])

    # Put label number
    cv2.putText(output, str(label), (x, y),
                cv2.FONT_HERSHEY_SIMPLEX,
                0.7, (255, 255, 255), 2)

# Step 5: Display
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.title("Original Image")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.subplot(1,2,2)
plt.title(f"Labeled Components: {num_labels-1}")
plt.imshow(cv2.cvtColor(output, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.tight_layout()
plt.show()

print("Number of Objects Detected:", num_labels - 1)
```
