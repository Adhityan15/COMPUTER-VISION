```python
import cv2
import matplotlib.pyplot as plt

# Read the input image
img = cv2.imread("test.png")

if img is None:
    print("Image not loaded")
    exit()

# Resize for better display
img = cv2.resize(img, (512, 512))

# Define text properties
text = "OpenCV Text Annotation"
position = (50, 100)              # (x, y)
font = cv2.FONT_HERSHEY_SIMPLEX
font_scale = 1
color = (0, 0, 255)               # Red color (BGR)
thickness = 2
line_type = cv2.LINE_AA

# Put text on image
annotated_img = cv2.putText(
    img,
    text,
    position,
    font,
    font_scale,
    color,
    thickness,
    line_type
)

# Display output using matplotlib
plt.imshow(cv2.cvtColor(annotated_img, cv2.COLOR_BGR2RGB))
plt.title("Text Annotation using OpenCV")
plt.axis("off")
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Create a blank white image
img = np.ones((400, 600, 3), dtype=np.uint8) * 255

# Draw a rectangle
cv2.rectangle(
    img,
    (50, 50),        # top-left corner
    (250, 200),      # bottom-right corner
    (0, 0, 0),       # black color
    2                # thickness
)

# Draw a circle
cv2.circle(
    img,
    (400, 120),      # center
    70,              # radius
    (0, 0, 0),       # black color
    2
)

# Draw a line
cv2.line(
    img,
    (100, 300),      # start point
    (500, 300),      # end point
    (0, 0, 0),       # black color
    2
)

# Display output
plt.imshow(img)
plt.title("Shape-Based Image Annotation")
plt.axis("off")
plt.show()
```
